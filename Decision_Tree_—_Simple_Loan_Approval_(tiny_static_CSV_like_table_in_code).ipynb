# **Decision Tree — Simple Loan Approval (tiny static CSV-like table in code)**

```
from typing import Optional
#Decision Tree — Simple Loan Approval (tiny static CSV-like table in code)
# Real-life idea: approve/deny a loan with simple business rules learned from data
# We’ll build a tiny static dataset in code (no files)
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Features: [income_k, credit_score, years_at_job, has_previous_default]
data = np.array([
    [35, 710, 2, 0],
    [18, 650, 1, 0],
    [55, 720, 8, 0],
    [25, 600, 0, 1],
    [80, 780, 10, 0],
    [45, 690, 3, 0],
    [20, 590, 1, 1],
    [60, 730, 5, 0],
    [28, 640, 2, 0],
    [50, 700, 7, 0],
    [22, 610, 0, 1],
    [40, 675, 4, 0],
])

# Labels: 1 = approve, 0 = deny (hand-crafted for demo)
labels = np.array([1,0,1,0,1,1,0,1,0,1,0,1])

Xtr, Xte, ytr, yte = train_test_split(data, labels, test_size=0.25, random_state=42, stratify=labels)

tree = DecisionTreeClassifier(max_depth=3, random_state=42)
tree.fit(Xtr, ytr)

pred = tree.predict(Xte)
print("Decision Tree accuracy:", accuracy_score(yte, pred))

# Try a new applicant
new_applicant = [[42, 705, 4, 0]]
print("Approve(1)/Deny(0):", tree.predict(new_applicant))




Optional
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plot_tree(tree, feature_names=["income_k","credit_score","years_at_job","prev_default"], class_names=["deny","approve"], filled=True)
plt.show()
```
